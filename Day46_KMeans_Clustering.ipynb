## Learnings

. Unsupervised Mindset  
    . No predefined labels or "correct answers"  
    . Goal: maximize inter-cluster differences and intra-cluster similarity  

. Importance of Scaling  
    . Larger-value features dominate distance calculations  
    . Standardization is essential for fair clustering  

. Elbow Method & Inertia  
    . Inertia = Within-Cluster Sum of Squares (WCSS)  
    . Helps measure cluster compactness  
    . Optimal K is found at the "elbow point"  

. Iterative Convergence  
    . K-Means follows an iterative process (Lloyd’s Algorithm)  
    . Centroids update until stable (convergence)  


## Blockers & Challenges

. Flat Elbow Problem  
    . Smooth curve made it hard to identify optimal K  
    . Cause:  
        . Non-spherical clusters  
        . Noisy data  
    . Insight: K-Means assumes spherical clusters  

. Initial Centroid Placement  
    . Poor initialization leads to bad clustering  
    . Solution:  
        . Use `init='k-means++'` for better centroid selection
